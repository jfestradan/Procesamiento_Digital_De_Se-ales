```python
%matplotlib inline
import numpy as np
import  matplotlib.pyplot as plt
N = 30 # numero de trayectorias, intentos, epocas (simulaciones de monte carlo)
T = 100 # numero de muestras  por trayectoria
sigma = [0.1,1,10] # varianza del ruido blanco gaussiano
X = np.sqrt(sigma[1])*np.random.randn(N,T)
Y = np.sqrt(sigma[1])*np.random.randn(N,T)  
Z = np.sqrt(sigma[1])*np.random.randn(N,T)  
plt.plot(X.T)
plt.title('Ruido blanco gaussiano 0.1')
plt.show()
plt.plot(Y.T)
plt.title('Ruido blanco gaussiano 1')

plt.show()
plt.plot(Z.T)
plt.title('Ruido blanco gaussiano 10')

plt.show()
Rx  = (1/N)*X.T @ X #matriz de correlaciones Rxx(tn,tm) , matriz de tama√±o T x T (X.T).dot(X)
plt.plot(X.mean(axis=0),label='media_ens N') # media por filas - vector de T
plt.plot(X.mean(axis=1),label = 'media_t') #media por columnas - vector de N
plt.plot(X.var(axis=0), label='var_ens N')
plt.plot(X.var(axis=1), label = 'var_por t')
plt.plot(np.diag(Rx), label = 'var diag Rx')
plt.legend()
plt.show()
plt.imshow(Rx)
plt.colorbar()
plt.show()

Ry  = (1/N)*Y.T @ Y 
plt.plot(Y.mean(axis=0),label='media_ens N') # media por filas - vector de T
plt.plot(Y.mean(axis=1),label = 'media_t') #media por columnas - vector de N
plt.plot(Y.var(axis=0), label='var_ens N')
plt.plot(Y.var(axis=1), label = 'var_por t')
plt.plot(np.diag(Ry), label = 'var diag Ry')
plt.legend()
plt.show()
plt.imshow(Ry)
plt.colorbar()
plt.show()

Rz  = (1/N)*Z.T @ Z
plt.plot(Z.mean(axis=0),label='media_ens N') # media por filas - vector de T
plt.plot(Z.mean(axis=1),label = 'media_t') #media por columnas - vector de N
plt.plot(Z.var(axis=0), label='var_ens N')
plt.plot(Z.var(axis=1), label = 'var_por t')
plt.plot(np.diag(Rz), label = 'var diag Rz')
plt.legend()
plt.show()
plt.imshow(Rz)
plt.colorbar()
plt.show()

print(X.mean(axis=0).mean())
print(X.mean(axis=1).mean())
print(X.var(axis=0).mean())
print(X.var(axis=1).mean())
```
